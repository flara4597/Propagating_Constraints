### Relationship between Farenheit and Celsius temperatures
- f = 9/5 * c + 32 or c = (f - 32) * 5/9
- A one directional function computes from one unit to another

```python
def celsius_to_farenheit(cel):
    """A one directional conversion from celsius to farenheit"""
    farenheit = 9 / 5 * cel + 32
    return farenheit
```

```python
def farenheit_to_celsius(far):
    """A one directional conversion from farenheit to celsius"""
    celsius = (far - 32) * (5 / 9)
    return celsius
```

```python
celsius_to_farenheit(25)
```

```python
farenheit_to_celsius(77)
```
